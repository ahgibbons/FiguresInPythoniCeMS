## Loading the data

Data stored in text (**csv**) or excel (**xlsx, xls**) files can be loaded using **pd.read_csv** or **pd.read_excel**. (Other file types can also be used).

Here are the paths to the data we will use. On Windows, begin the text string with **r** because the path separator **\\** is a special character. Or use **\\\\** instead. Python can also use **/** as the path separator.

(I assume the **¥** path separator on Japanese language computers is the same as **\\**? Not sure how this behaves ¯\\_(ツ)_/¯ )

### Loading a CSV file

```python
motion_path = r'data\dist_v_time.csv'
motion = pd.read_csv(motion_path)
motion
```

### Skipping rows

Sometimes the data does not begin at the beginning of the file. Sometimes there is other text (meta data) that should be **skipped**.

The **header** keyword lets you say which row is the header row (numbering starting from 0).

If your data does not have a header (why?) you can use **skiprows** keyword instead.

```python
spectra = pd.read_csv(r'data/spectra_wide.csv',header=1)
spectra

#spectra_wide.plot(x='Wavelength')
```

### Tab separated files

Some files separate by tabs or spaces instead of commas. Choose your separator using the **sep** keyword. The tab character is written **'\\t'**

```python
motion2_path = r'data\dist_v_time.txt'
motion2 = pd.read_csv(motion2_path, sep='\t')
motion2
```

## Loading Excel files

Excel files can be opened similar to csv files. If there are multiple sheets then the **sheet_name** keyword can be used.

**sheet_name = 1**  
Use a number to select excel sheet by number (starting with 0)

**sheet_name = "name of sheet"**  
Use text to select the sheet by name

**sheet_name = None**  
Special case, load **all** sheets as a **python dictionary**

```python
data1 = pd.read_excel('data/generated_data.xlsx', sheet_name=3)  # 0, 1, 2, *3*
data2 = pd.read_excel('data/generated_data.xlsx', sheet_name='coffee_and_happiness')
data_dict = pd.read_excel('data/generated_data.xlsx', sheet_name=None)
data3 = data_dict['coffee_and_happiness']

display(data1)
display(data2)
display(data3)
```

## Combining Sheets

If multiple excel sheets have the same columns, you can join them together, as shown in this example.

```python
dist_sheets = pd.read_excel('data/Example_join_sheets.xlsx', sheet_name=None)
```

```python
dist_sheets
```

```python
dist = pd.concat(dist_sheets)


dist = dist.reset_index()

dist = dist.drop('level_1', axis=1)

dist = dist.rename(columns={'level_0' : 'Sample'})
"""
"""
dist
```

We can chain these operations together to make a handy one-liner!

```python
dist = pd.concat(dist_sheets).reset_index().drop('level_1',axis=1).rename(columns={'level_0' : 'Sample'})
dist
```

```python
sns.lineplot(data=dist, x='Time', y='Distance', hue='Sample')
```
