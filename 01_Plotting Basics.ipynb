## Very First Plot - Pyplot style

This example uses the **pyplot style** (plt.plot, plt.xlabel, plt.title, etc).

This is good for single, simple plots, without advanced editing.

*Ref: https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.plot.html*

```python
xs = np.linspace(0,20,100)       # 100 points between 0 and 20
#print(xs)


ys1 = np.sin(xs) 
plt.plot(xs,ys1)

ys2 = np.cos(xs)
plt.plot(xs,ys2)

plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Sine and Cosine')

plt.scatter(xs,ys1)
"""
""";
```

## Very First Plot - Object / Axes style

This example uses the **Axes style** of plotting (ax.plot, ax.set_xlabel, ax.set_title, etc).



This style is good for making more advanced figures and figures with multiple Axes.

*Ref:* 
1. https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.subplots.html
2. https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.axes.Axes.plot.html


```python
fig,ax = plt.subplots()

xs = np.linspace(0,20,100)       # 100 points between 0 and 10
# print(xs)

ys1 = np.sin(xs)

ax.plot(xs,ys1)
ax.set_xticks([0,5,10,15,20])

ys2 = np.cos(xs)
ax.plot(xs,ys2,linestyle='--', color='r', marker='x')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sine and Cosine')


ax.legend(['Sine', 'Cosine'])

#plt.savefig('sine_cosine.png', dpi=300)
```

### Pyplot vs Axes Style?

Both styles can be used. **Axes** style is usually preferred when:

* There are multiple plots
* You are doing precise editing of the figure


# Seaborn (sns)

**Seaborn** is a library for quickly making nice figures. Works best with data loaded with **Pandas** (next part).
The name is usually shortened to **sns** (joke name from library author). 

It is good to start with **Seaborn**, then do further editing.

```python
import seaborn as sns             # Fancy plotting libary, renamed to 'sns'

sns.lineplot(x=xs,y=ys1)
ax = sns.lineplot(x=xs,y=ys2)

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sine and Cosine')

ax.legend(['Sine', 'Cosine'])
```

## Figure Dimensions and DPI

The figure dimensions are set by **figsize** and are in inches. The DPI can be set when the figure is created or when it is saved.

Academic journals usually require greater than **300 dpi** for images, and greater than **600 dpi** for line art.

The size of figures depends on the journal, but typical sizes are: 3.5 inch width (half column) and 7 inch (double column).

* **Science**: 3.54, 5, 7.25 inch width
* **Nature**: 3.5, 7.2 inch width

```python
## Half Page width figure

fig,ax = plt.subplots(figsize=(3.5,3))
sns.lineplot(x=xs,y=ys1)
ax = sns.lineplot(x=xs,y=ys2)

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sine and Cosine')

ax.legend(['Sine', 'Cosine'])

## Full page width figure

fig,ax = plt.subplots(figsize=(7.2,4))
sns.lineplot(x=xs,y=ys1)
ax = sns.lineplot(x=xs,y=ys2)

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sine and Cosine')

ax.legend(['Sine', 'Cosine'])
"""
""";
```

## Saving Figure

Figures can be saved using **plt.savefig(filename)** or **fig.savefig(filename)**. 

Supported formats, on my version of Matplotlib, are:

* **eps: Encapsulated Postscript**
* jpg: Joint Photographic Experts Group
* jpeg: Joint Photographic Experts Group
* **pdf: Portable Document Format**
* pgf: PGF code for LaTeX
* **png: Portable Network Graphics**
* ps: Postscript
* raw: Raw RGBA bitmap
* rgba: Raw RGBA bitmap
* **svg: Scalable Vector Graphics**
* svgz: Scalable Vector Graphics
* tif: Tagged Image File Format
* **tiff: Tagged Image File Format**

Common useful formats highlighted in **bold**. JPG is not recommended for text and lineart images.

```python
fig,ax = plt.subplots(figsize=(7.2,4))


ax.plot(xs,ys1)
ax.plot(xs,ys2,linestyle='--', color='r', marker='x')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Sine and Cosine')

ax.legend(['Sine', 'Cosine'], loc='lower right')
#plt.savefig("sine_cosine.png", dpi=600)
fig.savefig("sine_cosine.png", dpi=600)
fig.savefig("sine_cosine.pdf", dpi=600)
```

## Colors

There is a default set of colors used. They can be referenced by using "C1", "C2", "C3"...."C9".

Other colors can be specified.

```python
fig,ax = plt.subplots()

xs = np.linspace(0,20,100) 
for i in range(10):
    ys = np.sin(xs/(i+1))
    ax.plot(xs,ys, label=f'C{i}')
    
ax.legend(bbox_to_anchor=[1,0.2],loc='lower left')
```

Colors can be specifically chosen using hexcode (example #f0b223), or rgb fraction lists, (0.7, 0.2, 1.0) 

```python
fig,ax = plt.subplots()

xs = np.linspace(0,20,100)
ys1 = np.sin(xs)
ys2 = np.cos(xs)
ys3 = np.cos(xs)*np.cos(xs)

ax.plot(xs,ys1, color='#bc19e4', label='purple')
ax.plot(xs,ys2, color=(0.8, 0.6, 0.2), label='goldenrod?')
ax.plot(xs,ys3, color='blue', label='blue')
ax.legend()
```
