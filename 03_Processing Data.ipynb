# Worked Example, Approach one

Here is a step by step example of processing some data, and plotting it.

```python
dist = pd.read_csv(r'data\dist_v_time.csv')
dist
```

(plotting directly using pyplot)

```python
fig,ax = plt.subplots()
ax.plot(dist['Time'], dist['S1'], label='S1', marker='o')
ax.plot(dist['Time'], dist['S2'], label='S2', marker='o')
ax.plot(dist['Time'], dist['S3'], label='S3', marker='o')
ax.plot(dist['Time'], dist['S4'], label='S4', marker='o')
ax.legend()
ax.set_xlabel('Time')
ax.set_ylabel('Distance')
```

(plotting using Dataframe)

```python
ax = dist.plot(x='Time', y=['S1', 'S2', 'S3', 'S4'], marker='o')
ax.set_ylabel('Distance')
```

## Wide to Tall
It is often convenient to have the data in tall form. Use the **melt** function to do this.

**melt** can be difficult to remember how to use, hopefully this example helps

```python
dist_tall = dist.melt(id_vars='Time',  value_vars=['S1','S2','S3','S4'], var_name='Sample', value_name='Distance')
#help(dist.melt)
display(dist)
dist_tall
```

```python
sns.lineplot(data=dist_tall, x='Time', y='Distance', hue='Sample', marker='o')
```

## Adding new columns

Tall form data is convenient for adding new columns

```python
dist_tall['Dist. Square'] = np.power(dist_tall['Distance'],2)
dist_tall
```

```python
sns.lineplot(data=dist_tall, x='Time', y='Dist. Square', hue='Sample')
```

```python
fig,axs = plt.subplots(ncols=2, figsize=(10,4))

sns.lineplot(data=dist_tall, x='Time', y='Dist. Square', hue='Sample', ax=axs[0])
sns.lineplot(data=dist_tall, x='Time', y='Distance', hue='Sample', ax=axs[1])
```
